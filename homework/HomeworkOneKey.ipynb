# Q3

Which type of join will return the smallest subset of data, and why?

Answer: An inner join combines two DataFrames based on a join key and returns a new DataFrame that contains only those rows that have matching values in both of the original DataFrames. This will be the smallest data subset because a value must be in _both_ datasets to be included.

### Q5 

Why do we need yearly_data_csv_writer()? Why not just split up the raw file?
Answer: Modifying the raw data means we could introduce errors. Instead, we want to save our changes to a new file. That way, if there is ever any doubt about what we did to the data, and how, we can look back and see when changes were introduced.
